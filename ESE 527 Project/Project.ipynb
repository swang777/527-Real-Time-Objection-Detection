```python
from pycocotools.coco import COCO
import cv2
import os

# Initialize COCO API for instance annotations
annotations_path = "coco/annotations/instances_train2017.json"
coco = COCO(annotations_path)

# Directory for the images
img_dir = "coco/images/train2017/"
output_dir = "processed_images/"
os.makedirs(output_dir, exist_ok=True)

# Get all image IDs
imgIds = coco.getImgIds()

for img_id in imgIds:
    # Load image
    img_info = coco.loadImgs(img_id)[0]
    img_path = os.path.join(img_dir, img_info['file_name'])
    img = cv2.imread(img_path)

    # Resize image
    resized_img = cv2.resize(img, (224, 224))  # Example: Resizing to 224x224

    # Save the processed image
    output_path = os.path.join(output_dir, img_info['file_name'])
    cv2.imwrite(output_path, resized_img)
```
